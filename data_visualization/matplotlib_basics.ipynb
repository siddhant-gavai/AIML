# Introduction to Data Visualization with Matplotlib

Matplotlib is the foundational visualization library in Python. It provides the ability to generate publication-quality figures, charts, and plots from arrays or DataFrames.

This tutorial covers:
1. **Line Plots** (and customizing colors, markers, and line styles)
2. **Bar Charts** (plotting categorical data)
3. **Scatter Plots** (visualizing relationships between numeric variables)
4. **Subplots** (creating multi-panel layouts)

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Creating a Basic Line Plot

Let's plot a mathematical curve (sine wave) and add a title, axes labels, a grid, and custom line aesthetics.

```python
# Generate mock data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, color='darkblue', linestyle='--', linewidth=2, label='Sine Wave')

# Customize plot
plt.title('Sine Wave Curve', fontsize=14, fontweight='bold')
plt.xlabel('X-Axis Values', fontsize=12)
plt.ylabel('Y-Axis Values (sin)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='upper right')

# Show plot
plt.show()
```

## 2. Categorical Bar Charts

Bar charts are excellent for comparing metrics across distinct categories. Let's plot product sales.

```python
products = ['Laptops', 'Desktops', 'Monitors', 'Keyboards', 'Mice']
sales = [120, 85, 150, 200, 180]

plt.figure(figsize=(8, 4.5))
plt.bar(products, sales, color='teal', edgecolor='black', alpha=0.85)

plt.title('Product Sales Comparison', fontsize=14, fontweight='bold')
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

## 3. Scatter Plots

Scatter plots are perfect for analyzing correlations or patterns between two numerical variables. We can also add size and color variations based on a third variable.

```python
# Generate random data
np.random.seed(42)
study_hours = np.random.uniform(1, 10, 50)
exam_scores = 10 * study_hours + np.random.normal(0, 5, 50)
attendance = np.random.uniform(50, 100, 50)

plt.figure(figsize=(8.5, 5))
scatter = plt.scatter(study_hours, exam_scores, c=attendance, cmap='viridis', s=attendance*1.5, alpha=0.8, edgecolor='black')

plt.title('Relationship Between Study Hours and Exam Scores', fontsize=14, fontweight='bold')
plt.xlabel('Study Hours', fontsize=12)
plt.ylabel('Exam Scores', fontsize=12)

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Attendance Rate (%)', fontsize=11)

plt.show()
```

## 4. Multi-Panel Subplots

Using `plt.subplots()`, we can create a grid layout of multiple distinct plots inside a single figure.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4.5))

# First subplot: Sine wave
axes[0].plot(x, np.sin(x), color='red', label='Sine')
axes[0].set_title('Sine Function')
axes[0].grid(True)
axes[0].legend()

# Second subplot: Cosine wave
axes[1].plot(x, np.cos(x), color='green', label='Cosine')
axes[1].set_title('Cosine Function')
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()  # Adjusts spacing to prevent overlap
plt.show()
```
