## Filtering using the `query()` Method

The `query()` method allows you to filter a DataFrame using a boolean expression represented as a string. It often makes the code more readable and concise, especially for complex filtering conditions.

```python
# Filter movies with IMDb rating > 8.0 using query()
df.query("IMDb > 8.0")
```

```python
# Filter Action movies with Box Office > 500 Crore using query()
# Column names with spaces or special characters (like 'BoxOffice(INR Crore)') must be enclosed in backticks (`)
df.query("Genre == 'Action' and `BoxOffice(INR Crore)` > 500")
```

```python
# Filter movies referencing a local variable with the '@' symbol
min_rating = 7.5
df.query("IMDb > @min_rating")
```

```python
# Filter movies where the Genre is in a specific list
target_genres = ["Action", "Thriller"]
df.query("Genre in @target_genres")
```
