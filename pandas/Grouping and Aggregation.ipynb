# Pandas Grouping and Aggregations

Grouping data and calculating aggregations is one of the most common tasks in data analysis (similar to SQL's `GROUP BY`). Pandas provides the split-apply-combine workflow using the `.groupby()` method.

This tutorial covers:
1. **Basic Grouping** (`groupby()`)
2. **Built-in Aggregation Functions** (`sum()`, `mean()`, `count()`)
3. **Grouping by Multiple Columns**
4. **Custom Aggregations** with `.agg()`

```python
import pandas as pd
import numpy as np
```

## 1. Creating a Sample Sales Dataset

Let's create a DataFrame representing sales transactions across multiple categories and regions.

```python
data = {
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Office Supplies', 'Furniture', 'Electronics', 'Office Supplies'],
    'Region': ['East', 'West', 'West', 'East', 'East', 'East', 'West'],
    'Sales': [250.0, 350.0, 120.0, 15.0, 450.0, 300.0, 45.0],
    'Quantity': [2, 1, 4, 10, 3, 1, 5]
}
df = pd.DataFrame(data)
df
```

## 2. Basic Grouping and Aggregating

To compute aggregate metrics per Category, we call `groupby('Category')` followed by a summary function. For instance, let's find the **total Sales** per Category.

```python
# Group by Category and sum only the 'Sales' column
df.groupby('Category')[['Sales']].sum()
```

We can calculate the average (mean) Sales and Quantity per Category:

```python
# Compute mean for all numerical columns in the group
df.groupby('Category')[['Sales', 'Quantity']].mean()
```

## 3. Grouping by Multiple Columns

We can group by multiple columns by passing a list to `.groupby()`. This creates a **MultiIndex** DataFrame.

```python
# Group by both Category and Region
multi_group = df.groupby(['Category', 'Region']).sum()
multi_group
```

To convert a MultiIndex back to standard columns, call `.reset_index()`.

```python
multi_group.reset_index()
```

## 4. Custom Aggregations using `.agg()`

The `.agg()` method offers ultimate flexibility, allowing you to:
- Apply multiple statistical functions to a single column.
- Apply different functions to different columns.

```python
# Perform multiple aggregations on Sales, and find max on Quantity
df.groupby('Category').agg({
    'Sales': ['sum', 'mean'],
    'Quantity': 'max'
})
```
