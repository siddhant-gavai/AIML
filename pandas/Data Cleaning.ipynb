# Pandas Data Cleaning and Handling Missing Values

In real-world data analysis, datasets are rarely clean. They often contain missing values, placeholder values, or incorrect data types. Pandas provides a powerful and flexible set of tools to identify, clean, and resolve these issues.

This tutorial covers:
1. **Detecting Missing Values** (`isna()`, `isnull()`, `.sum()`)
2. **Dropping Missing Data** (`dropna()`)
3. **Filling Missing Data** (`fillna()`, forward fill, backward fill)
4. **Replacing Values** (`replace()`)

```python
import pandas as pd
import numpy as np
```

## 1. Detecting Missing Values

Let's create a sample DataFrame representing student records, with some missing values (`None` or `np.nan`).

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [20.0, np.nan, 22.0, 21.0, 19.0],
    'Score': [85.0, 90.0, np.nan, 88.0, np.nan],
    'City': ['New York', None, 'Boston', 'Chicago', None]
}
df = pd.DataFrame(data)
df
```

To detect missing values, we use `.isna()` (or its alias `.isnull()`). This returns a boolean DataFrame of the same shape.

```python
df.isna()
```

To count the number of missing values in each column, chain it with `.sum()`.

```python
df.isna().sum()
```

## 2. Dropping Missing Data

We can use `.dropna()` to remove rows or columns containing missing values.

By default, `dropna()` drops any row that has at least one missing value (`how='any'`).

```python
# Drop rows with any NaN values
df.dropna()
```

To drop columns instead of rows, use `axis=1` or `axis='columns'`.

```python
# Drop columns with any NaN values - since 'Name' is the only fully populated column, we select only it.
# Note: df is not modified unless inplace=True is set.
df.dropna(axis=1)
```

## 3. Filling Missing Data

Instead of discarding data, we can fill missing values using `.fillna()`.

```python
# Fill numerical NaNs with 0 and strings with 'Unknown'
df.fillna({'Age': 0, 'Score': 0, 'City': 'Unknown'})
```

We can also fill missing values dynamically with summary statistics like the **mean** or **median** of the column.

```python
# Fill Age and Score with their respective mean values
df_clean = df.copy()
df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].mean())
df_clean['Score'] = df_clean['Score'].fillna(df_clean['Score'].mean())
df_clean
```

### Forward Fill and Backward Fill

For sequential or time-series data, we can propagate non-null values forward (`ffill`) or backward (`bfill`).

```python
# Forward fill propagates the last valid observation down to next valid
df.ffill()
```

## 4. Replacing Values

Sometimes, missing values or placeholders are written as specific characters (like `"?"`, `"N/A"`, or `-999`). We can use `.replace()` to swap them out.

```python
df_raw = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Price': [1200, -999, 80, '?']
})

# Replace -999 and '?' with NaN to mark them as missing values
df_clean_prices = df_raw.replace([-999, '?'], np.nan)
df_clean_prices
```
