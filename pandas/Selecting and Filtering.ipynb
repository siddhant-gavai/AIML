# Indexing, Selecting, and Filtering Data in Pandas

Once you have loaded or created a DataFrame, you need to extract specific rows, columns, or values. Pandas provides extremely powerful indexing and filtering mechanisms.

## Setup a Sample DataFrame

Let's create a DataFrame containing student scores and details to use for our selections.

```python
import pandas as pd

student_data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [20, 21, 19, 22, 20],
    "Math": [85, 90, 78, 65, 95],
    "Science": [90, 85, 88, 72, 98],
    "Passed": [True, True, True, False, True]
}
df = pd.DataFrame(student_data)
df
```

## 1. Selecting Columns

* **Single Column (Series)**: Use single brackets `df['ColumnName']`.
* **Multiple Columns (DataFrame)**: Use a list of column names inside brackets `df[['Col1', 'Col2']]`.

```python
# Select a single column as a Series
math_scores = df["Math"]
print("Math Scores Series:\n", math_scores)

# Select multiple columns as a DataFrame
grades = df[["Name", "Math", "Science"]]
print("\nGrades DataFrame:\n", grades)
```

## 2. Row Selection using `.loc` and `.iloc`

* **`.loc[]`**: Label-based index selector (selects by row labels/indexes).
* **`.iloc[]`**: Integer-position-based selector (selects by integer positions, starting from 0).

```python
# Set "Name" as index to demonstrate label-based selection better
df_named = df.set_index("Name")
print("Named DataFrame:\n", df_named)

# 1. Label-based selection (.loc)
print("\n--- Row for Bob (.loc) ---")
print(df_named.loc["Bob"])

# Select multiple specific rows and columns
print("\n--- Specific cells for Alice and Charlie ---")
print(df_named.loc[["Alice", "Charlie"], ["Math", "Science"]])

# 2. Integer-position-based selection (.iloc)
print("\n--- First Row (.iloc[0]) ---")
print(df_named.iloc[0])

# Select slice of rows and columns (rows 1 to 3, columns 0 to 2)
print("\n--- Row slice 1:4, Column slice 0:2 (.iloc) ---")
print(df_named.iloc[1:4, 0:2])
```

## 3. Conditional / Boolean Filtering

You can filter rows by evaluating conditional expressions. This is one of the most powerful features of Pandas.

```python
# Find students who scored above 80 in Math
high_math = df[df["Math"] > 80]
print("Students with Math > 80:\n", high_math)

# Complex logical conditions
# Use & for AND, | for OR, and ~ for NOT. Wrap each condition in parentheses!
failed_or_low_science = df[(df["Passed"] == False) | (df["Science"] < 80)]
print("\nStudents who failed OR have Science score < 80:\n", failed_or_low_science)
```
