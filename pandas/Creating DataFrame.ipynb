# Creating DataFrames in Pandas

A **DataFrame** is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or a SQL table.

Let's explore 5 different ways to create a DataFrame from scratch.

## 1. Creating a DataFrame from a Dictionary of Lists

This is the most common way to construct a DataFrame. Each key in the dictionary becomes a column, and the list associated with the key represents the values for that column.

```python
import pandas as pd
import numpy as np

data_dict = {
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor"],
    "Price": [1200, 25, 75, 300],
    "In_Stock": [True, True, False, True]
}
df1 = pd.DataFrame(data_dict)
df1
```

## 2. Creating a DataFrame from a List of Dictionaries

This approach is useful when handling record-oriented data, such as responses from JSON APIs or database queries.

```python
records = [
    {"Name": "Alice", "Role": "Engineer", "Salary": 95000},
    {"Name": "Bob", "Role": "Designer", "Salary": 85000},
    {"Name": "Charlie", "Role": "Manager", "Salary": 110000}
]
df2 = pd.DataFrame(records)
df2
```

## 3. Creating a DataFrame from a List of Lists (with explicit columns)

When creating a DataFrame from a nested list, you need to explicitly supply the column names using the `columns` parameter.

```python
list_data = [
    ["London", 8.9, "UK"],
    ["Paris", 2.1, "France"],
    ["New York", 8.3, "USA"]
]
col_names = ["City", "Population_M", "Country"]
df3 = pd.DataFrame(list_data, columns=col_names)
df3
```

## 4. Creating a DataFrame from a NumPy Array

If you already have data in a NumPy array, you can easily load it into a DataFrame and name the features.

```python
random_data = np.random.randn(5, 3)
df4 = pd.DataFrame(random_data, columns=["Feature A", "Feature B", "Feature C"])
df4
```

## 5. Setting and Resetting a Custom Index

By default, Pandas assigns integer indexes (0, 1, 2, ...). You can set any column to be the primary index, which is highly useful for labels.

```python
# Set "Product" as the row index
df_indexed = df1.set_index("Product")
print("Indexed DataFrame:\n", df_indexed)

# Resetting the index back to default integers
print("\nReset Index:\n", df_indexed.reset_index())
```
