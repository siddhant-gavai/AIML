```python
import numpy as np
import time
```

```python
#execution perfomance
size = 10_000_000

#python list 
py_list =  list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time = {end-start} seconds")

np_array =  np.array(py_list)
start = time.time()

sq_array = np_array ** 2 
end = time.time()
print(f"numpy array  time = {end-start} seconds")
```

```python
#memory 
import sys

print(f"python list size = {sys.getsizeof(py_list) * len(py_list)} bytes ")
print(f"numpy array size = {np_array.nbytes} bytes ")
```

Memory Performance
Comparing memory consumption of Python lists vs NumPy arrays.

```python
import sys

# Memory consumption of a Python list
py_list_size = 1000
py_list_mem = list(range(py_list_size))
print(f"Size of each element in Python list: {sys.getsizeof(py_list_mem[0])} bytes")
print(f"Total size of Python list: {sys.getsizeof(py_list_mem[0]) * len(py_list_mem)} bytes (approx)")

# Memory consumption of a NumPy array
np_array_mem = np.arange(py_list_size)
print(f"Size of each element in NumPy array: {np_array_mem.itemsize} bytes")
print(f"Total size of NumPy array: {np_array_mem.nbytes} bytes")
```
