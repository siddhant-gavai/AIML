## NumPy Vectorization

**Vectorization** is one of the most powerful features of NumPy. It refers to the ability to perform element-wise operations on arrays without using explicit Python `for` loops.

### Why is Vectorization so fast?
1. **Compiled C Backend:** Operations are offloaded to pre-compiled, highly optimized C code under the hood.
2. **No Dynamic Type Checking:** Python loops check types and look up methods at *each* iteration. NumPy arrays are homogeneous (all elements have the same dtype), allowing compiled code to run continuously without checks.
3. **SIMD & Memory Contiguity:** NumPy uses contiguous blocks of memory, allowing CPUs to leverage SIMD (Single Instruction, Multiple Data) instructions to process multiple data points in a single clock cycle.

Let's see a practical comparison between standard Python loops and NumPy vectorization.

```python
import time

# Create a list and a NumPy array with 1,000,000 elements
size = 1_000_000
py_list = list(range(size))
np_arr = np.arange(size)

# Goal: Add 5 to each element and measure time

# 1. Traditional Python Loop approach
start_time = time.time()
py_list_result = [x + 5 for x in py_list]
py_duration = time.time() - start_time
print(f"Python List (List Comprehension) took: {py_duration:.6f} seconds")

# 2. NumPy Vectorized approach
start_time = time.time()
np_arr_result = np_arr + 5
np_duration = time.time() - start_time
print(f"NumPy Vectorized addition took:      {np_duration:.6f} seconds")

# Calculate speedup
speedup = py_duration / np_duration
print(f"Vectorized operation is ~{speedup:.1f}x faster!")
```

### Example 2: Vectorized Operations between Two Arrays

We can perform element-wise arithmetic, comparisons, and mathematical functions directly on entire arrays.

```python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

# Vectorized element-wise addition
print("a + b =", a + b)

# Vectorized multiplication
print("a * b =", a * b)

# Vectorized comparison (returns a boolean array)
print("b > 25 =", b > 25)

# Vectorized mathematical functions (using NumPy ufuncs)
print("sin(a) =", np.sin(a))
print("exp(a) =", np.exp(a))
```

## Vector Normalization

**Vector Normalization** (often referred to as L2 or Euclidean normalization) scales a non-zero vector so that its magnitude (norm or length) becomes exactly $1$. This is a crucial preprocessing step in machine learning, deep learning, and computer vision (e.g., when working with cosine similarity, word embeddings, or neural network inputs).

### Mathematical Definition:
For a vector $v$, the normalized vector $\hat{v}$ is defined as:
$$\hat{v} = \frac{v}{\|v\|_2}$$
where $\|v\|_2$ is the Euclidean (L2) norm of $v$:
$$\|v\|_2 = \sqrt{v_1^2 + v_2^2 + ... + v_n^2}$$

Let's implement vector normalization in NumPy using both a manual vectorized approach and the built-in linear algebra module (`np.linalg.norm`).

```python
# 1. Normalizing a 1D Vector
v = np.array([3.0, 4.0, 0.0]) # A 3D spatial vector
print("Original Vector (v):", v)

# Method A: Manual vectorized calculation
l2_norm_manual = np.sqrt(np.sum(v**2))
v_normalized_manual = v / l2_norm_manual
print("\nMethod A (Manual L2 Normalization):")
print("L2 Norm (Magnitude):", l2_norm_manual)
print("Normalized Vector:", v_normalized_manual)
print("Magnitude of Normalized Vector:", np.sqrt(np.sum(v_normalized_manual**2)))

# Method B: Using np.linalg.norm
l2_norm_linalg = np.linalg.norm(v)
v_normalized_linalg = v / l2_norm_linalg
print("\nMethod B (using np.linalg.norm):")
print("L2 Norm (Magnitude):", l2_norm_linalg)
print("Normalized Vector:", v_normalized_linalg)
print("Magnitude of Normalized Vector:", np.linalg.norm(v_normalized_linalg))
```

### Normalizing Multiple Vectors in a Matrix (2D Array)

In machine learning, we often have a matrix where each row (or column) represents a separate feature vector. We want to normalize each row (or column) independently. 

We can do this in a single vectorized statement by utilizing the `axis` parameter and **broadcasting**.

```python
# A matrix containing three 2D vectors (each row is a vector)
matrix = np.array([
    [3.0, 4.0],
    [1.0, 2.0],
    [0.0, 5.0]
])
print("Original Matrix:\n", matrix)

# Step 1: Calculate the L2 norm of each row (axis=1)
# We use keepdims=True to keep the dimensions as (3, 1) for broadcasting
row_norms = np.linalg.norm(matrix, axis=1, keepdims=True)
print("\nRow Norms (shape [3, 1]):\n", row_norms)

# Step 2: Normalize the matrix (each row is divided by its respective norm)
normalized_matrix = matrix / row_norms
print("\nNormalized Matrix (each row has magnitude 1.0):\n", normalized_matrix)

# Verify that each row now has a norm of 1.0
print("\nVerification (Norms of normalized rows):", np.linalg.norm(normalized_matrix, axis=1))
```
