## NumPy Vectorization

**Vectorization** is one of the most powerful features of NumPy. It refers to the ability to perform element-wise operations on arrays without using explicit Python `for` loops.

### Why is Vectorization so fast?
1. **Compiled C Backend:** Operations are offloaded to pre-compiled, highly optimized C code under the hood.
2. **No Dynamic Type Checking:** Python loops check types and look up methods at *each* iteration. NumPy arrays are homogeneous (all elements have the same dtype), allowing compiled code to run continuously without checks.
3. **SIMD & Memory Contiguity:** NumPy uses contiguous blocks of memory, allowing CPUs to leverage SIMD (Single Instruction, Multiple Data) instructions to process multiple data points in a single clock cycle.

Let's see a practical comparison between standard Python loops and NumPy vectorization.

```python
import time

# Create a list and a NumPy array with 1,000,000 elements
size = 1_000_000
py_list = list(range(size))
np_arr = np.arange(size)

# Goal: Add 5 to each element and measure time

# 1. Traditional Python Loop approach
start_time = time.time()
py_list_result = [x + 5 for x in py_list]
py_duration = time.time() - start_time
print(f"Python List (List Comprehension) took: {py_duration:.6f} seconds")

# 2. NumPy Vectorized approach
start_time = time.time()
np_arr_result = np_arr + 5
np_duration = time.time() - start_time
print(f"NumPy Vectorized addition took:      {np_duration:.6f} seconds")

# Calculate speedup
speedup = py_duration / np_duration
print(f"Vectorized operation is ~{speedup:.1f}x faster!")
```

### Example 2: Vectorized Operations between Two Arrays

We can perform element-wise arithmetic, comparisons, and mathematical functions directly on entire arrays.

```python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

# Vectorized element-wise addition
print("a + b =", a + b)

# Vectorized multiplication
print("a * b =", a * b)

# Vectorized comparison (returns a boolean array)
print("b > 25 =", b > 25)

# Vectorized mathematical functions (using NumPy ufuncs)
print("sin(a) =", np.sin(a))
print("exp(a) =", np.exp(a))
```
